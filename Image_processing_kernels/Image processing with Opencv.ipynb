## Resizing Images
Sometimes its necessay to resize the image to particular width or height. We can for example resize down a 1920X1080 to a 1280X720 image or resize it based on particular ratios.

```python
fix_image.shape
```

```python
#Resizing the image to 1000X400
new_image = cv2.resize(fix_image,(1000,400))
plt.imshow(new_image)
```

Sometimes its preferable to reduce by a particular ratio instead of hardcoded values of height and width. For example instead of resizing to 1000 as width and 400 as height, we would prefer to reduce the width by 50% and height by 20%.  

```python
#Resizing by ratios
w_ratio = 0.8 #Resizing to 80% of its width  
h_ratio = 0.4 #Resizing to 40% of its height
```

```python
new_image = cv2.resize(fix_image,(0,0),fix_image,w_ratio,h_ratio)
plt.imshow(new_image)
```

## Drawing Shapes and Inscribing Text on images

We can plot different shapes on images such as rectangles circles etc. This is the foundational block of a detection box around a text.

```python
#Pure black Blank image of 3 channels
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

Plotting a rectangle on an image involves a simple superposition of a rectangle on top of the image by plotting the the coordinates with `pt1`=(`xmin,ymin`) & `pt2`=(`xmax`,`ymax`) of a particular thickness.

```python
cv2.rectangle(blank_img,pt1=(10,10),pt2=(100,200),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

Note that this doesn't require to assign to a variable

```python
cv2.rectangle(blank_img,pt1=(400,10),pt2=(500,200),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(255,0,0),thickness=-1)
```

```python
plt.imshow(blank_img)
```

Every new shape just superimposes on the previosuly superimposed image.

```python
cv2.circle(blank_img,center=(250,410),radius=50,color=(0,0,255),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(200,410),pt2=(300,410),color=(102,255,255),thickness=5)
cv2.line(blank_img,pt1=(250,450),pt2=(250,350),color=(102,255,255),thickness=5)
```

```python
plt.imshow(blank_img)
```

Plotting rectangles,circles and lines are simple. But there are complexities when it comes to complex shapes like polygons and plotting text. These reuire additional steps before plotting the figures on the image. Lets plot a piece of text on an image given before.

```python
#Define the font of the text
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hello',org=(20,100),fontFace=font,fontScale=1,color=(200,200,100),thickness=5)
plt.imshow(blank_img)
```

```python
cv2.putText(blank_img,text='Hello',org=(410,100),fontFace=font,fontScale=1,color=(200,100,200),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

The fontScale is the multiplier of the base fontFace so you need to appropriately choose the right size. Now for custom polygons seem a little bit complex as they need to be certain considerations which need to put in place. First we need to find the coordinates of the custom polygon as a seperate list of vertices of the polygon.

```python
#Create a new blank black image
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,200],[300,100],[500,200],[300,400]],dtype=np.int32)
```

```python
vertices.shape
```

We can see that the shape of vertices is in 4,2 dimesnion. According to OpenCV, it only accepts images in 3 dimension i.e. it requires channel input as well. Hence we need to reshape the vertices array. While reshaping,we input -1 to retain the original dimensions while just changing the structure of the array.

```python
pts = vertices.reshape(-1,1,2)
```

```python
pts.shape
```

```python
pts
```

We have made it into 3D array while retaining the original dimensions. Now when we finally plot the coordinates, we need to add the `pts` into a list. 

```python
cv2.polylines(blank_img,pts=[pts],isClosed=True,color=(255,255,0),thickness=1)
plt.imshow(blank_img)
```
